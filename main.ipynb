### 2- Intégrale

Les approximations de Riemann, telles que celles illustrées par les sommes des rectangles à gauche et à droite, sont des outils fondamentaux en analyse numérique pour estimer les intégrales. Leur principal avantage réside dans leur simplicité conceptuelle et leur facilité de mise en œuvre, en particulier pour les intégrales qui ne peuvent pas être résolues analytiquement.

**Qualité des approximations :**
- **Sommes de gauche (Left Riemann Sum)** : Cette approximation est généralement plus précise lorsque la fonction est décroissante sur l'intervalle considéré. Cependant, elle peut sous-estimer l'aire si la fonction est croissante.
- **Sommes de droite (Right Riemann Sum)** : Inversement, cette approximation a tendance à être plus précise pour les fonctions croissantes, mais surévalue l'aire pour les fonctions décroissantes.

**Intérêts des approximations :**
- **Calculs Numériques** : Quand on ne peut pas intégrer une fonction analytiquement ou quand on a affaire à des données empiriques ou expérimentales.
- **Enseignement et Compréhension** : Elles fournissent une méthode visuelle intuitive pour comprendre les intégrales définies.
- **Algorithmes** : Ces méthodes sont la base de techniques d'intégration numérique plus sophistiquées comme la méthode des trapèzes ou la méthode de Simpson.

**Calcul des formules approchées de \( I(f) \) :**

Pour une fonction \( f \) continue sur l'intervalle \([a, b]\), l'intégrale approchée \( I(f) \) peut être calculée par les sommes de Riemann comme suit :

- **Somme de gauche (Left Riemann Sum)** :
  \[ I_{\text{left}}(f) = \sum_{i=0}^{N-1} f(t_i) \Delta t \]
  où \( \Delta t = \frac{b - a}{N} \), \( t_i = a + i\Delta t \), et \( N \) est le nombre de sous-intervalles.

-

Les formules approchées de l'intégrale \( I(f) \) pour la fonction \( f(t) = \cos(t) \) sur l'intervalle de \( 0 \) à \( 5 \) avec \( N = 4 \) sous-intervalles sont :

- Pour la **Somme de gauche (Left Riemann Sum)** : 
  \[ I_{\text{left}}(f) \approx -0.383 \]

- Pour la **Somme de droite (Right Riemann Sum)** : 
  \[ I_{\text{right}}(f) \approx -1.278 \]

Ces valeurs sont des approximations grossières car le nombre de sous-intervalles \( N \) est petit. Avec une valeur plus élevée de \( N \), les approximations seraient plus proches de la valeur réelle de l'intégrale. En général, la méthode des rectangles peut être particulièrement intéressante lorsque la fonction est difficile à intégrer analytiquement ou lorsque nous avons des données discrètes représentant la fonction.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate and the number of points
def f(t):
    return np.cos(t)

N = 4  # N+1 points, as per the user's request, here N is 4, hence 5 points

# Define the range for t
t_min, t_max = 0, 5
t = np.linspace(t_min, t_max, N+1)
f_t = f(t)

# Calculate the width of each rectangle (Delta t)
delta_t = (t_max - t_min) / N

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the function
ax.plot(t, f_t, 'r')

# Plot the rectangles for the left Riemann sum
for i in range(N):
    ax.add_patch(plt.Rectangle((t[i], 0), delta_t, f_t[i], edgecolor='blue', facecolor='blue', alpha=0.3))

# Plot the rectangles for the right Riemann sum
for i in range(1, N+1):
    ax.add_patch(plt.Rectangle((t[i-1], 0), delta_t, f_t[i], edgecolor='green', facecolor='green', alpha=0.3))

# Add some labels and a legend
ax.set_xlabel('t')
ax.set_ylabel('f(t)')
ax.set_title('Left and Right Riemann Sums')
ax.legend(['f(t)', 'Left Sum', 'Right Sum'])

# Display the plot
plt.show()
````

````python
# Let's calculate the approximate integral using Left and Right Riemann Sums for N=4
# We will use the same t and f(t) from the previous plot

# Approximate integral using the Left Riemann Sum
I_left = sum(f_t[:-1]) * delta_t

# Approximate integral using the Right Riemann Sum
I_right = sum(f_t[1:]) * delta_t

I_left, I_right
````

### Méthode point milieu

La méthode du point milieu est une autre technique d'intégration numérique qui tend à être plus précise que les méthodes des rectangles à gauche ou à droite. Elle est basée sur l'approximation de l'intégrale par la somme des aires de rectangles dont la hauteur est la valeur de la fonction au milieu de chaque sous-intervalle.

La formule de l'intégrale approchée \( I_{\text{mid}}(f) \) pour une fonction \( f \) sur un intervalle \([a, b]\) avec \( N \) sous-intervalles est donnée par :
\[ I_{\text{mid}}(f) = \sum_{i=0}^{N-1} f\left(a + \left(i + \frac{1}{2}\right)\Delta t\right) \Delta t \]
où \( \Delta t = \frac{b - a}{N} \) et le point \( a + \left(i + \frac{1}{2}\right)\Delta t \) est le milieu du \( i \)-ème sous-intervalle.

Pour la fonction \( f(t) = t^3 \), calculons l'intégrale approchée sur l'intervalle de 0 à 2 en utilisant la méthode du point milieu et comparons l'erreur avec les méthodes des rectangles gauche et droite que nous avons précédemment utilisées.

Le code Python pour la méthode du point milieu est le suivant :

La valeur approchée de l'intégrale de la fonction \( f(t) = t^3 \) sur l'intervalle de 0 à 2 en utilisant la méthode du point milieu avec \( N = 4 \) sous-intervalles est de \( I_{\text{mid}}(f) \approx 3.875 \).

La valeur exacte de l'intégrale, calculée analytiquement, est \( \frac{2^4}{4} = 4 \).

En comparant les erreurs par rapport à la valeur exacte de l'intégrale, nous obtenons :
- Erreur pour la Somme de gauche (Left Riemann Sum) : \( \approx 4.383 \)
- Erreur pour la Somme de droite (Right Riemann Sum) : \( \approx 5.278 \)
- Erreur pour la méthode du point milieu : \( \approx 0.125 \)

**Commentaire :**

La méthode du point milieu fournit une approximation nettement plus précise pour cette fonction particulière \( f(t) = t^3 \) dans cet intervalle par rapport aux méthodes des rectangles à gauche et à droite. Cela est dû au fait que la méthode du point milieu utilise la valeur de la fonction au milieu de chaque sous-intervalle, ce qui peut fournir une meilleure estimation moyenne de la fonction sur chaque sous-intervalle, surtout quand la fonction est régulière et bien comportée comme \( t^3 \). En général, pour les fonctions qui ont un comportement linéaire ou presque linéaire sur chaque sous-intervalle, la méthode du point milieu sera souvent plus précise que les méthodes des rectangles car elle compense mieux pour la courbure de la fonction.

````python
# Define the function t^3 and the number of subintervals
def f(t):
    return t**3

N = 4  # Number of subintervals
a, b = 0, 2  # Interval from 0 to 2
delta_t = (b - a) / N  # Width of each subinterval

# Calculate the midpoint sum
midpoints = a + (np.arange(N) + 0.5) * delta_t
I_mid = sum(f(midpoints)) * delta_t

# The actual integral value for comparison
actual_integral = (b**4 - a**4) / 4  # Integral of t^3 from 0 to 2

# Calculate the errors for each method
error_left = abs(I_left - actual_integral)
error_right = abs(I_right - actual_integral)
error_mid = abs(I_mid - actual_integral)

I_mid, actual_integral, error_left, error_right, error_mid
````

### Méthode des trapèzes

La méthode des trapèzes est une autre technique d'intégration numérique qui est plus précise que la méthode des rectangles car elle prend en compte la variation linéaire de la fonction sur chaque sous-intervalle. Plutôt que de supposer que la fonction est constante sur un sous-intervalle (comme dans les méthodes des rectangles gauche et droite), la méthode des trapèzes suppose que la fonction est affine, c'est-à-dire qu'elle peut être représentée comme une ligne droite entre les deux points extrêmes du sous-intervalle.

Pour un intervalle \([t_i, t_{i+1}]\), la fonction affine peut être exprimée comme \(f(t) = a_i t + b_i\). Les coefficients \(a_i\) et \(b_i\) sont déterminés par les valeurs de la fonction aux extrémités de l'intervalle:
 ```math
1. À \(t = t_i\), la fonction a une valeur \(f(t_i)\). Cela donne l'équation \(f(t_i) = a_i t_i + b_i\).
2. À \(t = t_{i+1}\), la fonction a une valeur \(f(t_{i+1})\). Cela donne l'équation \(f(t_{i+1}) = a_i t_{i+1} + b_i\).
```

En résolvant ce système de deux équations avec deux inconnues (\(a_i\) et \(b_i\)), on peut trouver les valeurs des coefficients pour la droite affine qui passe par les points \((t_i, f(t_i))\) et \((t_{i+1}, f(t_{i+1}))\).

La pente \(a_i\) de la fonction affine est donnée par la différence des valeurs de la fonction aux points \(t_i\) et \(t_{i+1}\), divisée par la largeur de l'intervalle:

\[ a_i = \frac{f(t_{i+1}) - f(t_i)}{t_{i+1} - t_i} \]

Une fois que nous avons \(a_i\), nous pouvons substituer soit dans l'équation pour \(t_i\) soit \(t_{i+1}\) pour trouver \(b_i\):

\[ b_i = f(t_i) - a_i t_i \]

ou

\[ b

\[ b_i = f(t_{i+1}) - a_i t_{i+1} \]


Ces deux dernières formules donneront la même valeur de \( b_i \) puisqu'elles sont basées sur la même ligne droite qui relie les deux points. En pratique, vous pouvez utiliser l'une ou l'autre pour calculer \( b_i \).

Avec \( a_i \) et \( b_i \) déterminés, la fonction affine \( f(t) = a_i t + b_i \) sur l'intervalle \([t_i, t_{i+1}]\) est complètement spécifiée. Vous pouvez alors intégrer cette fonction affine sur l'intervalle pour obtenir l'approximation de l'aire sous la courbe entre \( t_i \) et \( t_{i+1} \), qui sera l'aire d'un trapèze, d'où le nom de la méthode des trapèzes.

````python
# Assuming we already have the function f(t) defined from the previous examples,
# and the interval [a, b] is from 0 to 2, let's calculate the integral using the trapezoidal rule

# Define the interval and number of subintervals
a, b = 0, 2
N = 4  # We can change the number of subintervals for more accuracy
t = np.linspace(a, b, N+1)
f_t = f(t)
delta_t = (b - a) / N

# Trapezoidal Rule Sum
I_trap = (f_t[0] + f_t[-1] + 2 * sum(f_t[1:-1])) * (delta_t / 2)

# Calculate the error for the trapezoidal method
error_trap = abs(I_trap - actual_integral)

I_trap, error_trap
````

### Méthode de Simpson

La méthode de Simpson est une technique d'intégration numérique qui fournit une approximation précise de l'intégrale d'une fonction en supposant que la fonction peut être approximée par un polynôme de degré 2, ou parabole, sur chaque sous-intervalle. Elle utilise les polynômes de Lagrange pour interpoler la fonction en trois points : les deux extrémités de l'intervalle, \( t_i \) et \( t_{i+1} \), et le point milieu \( m_i \).

Le polynôme interpolateur de Lagrange \( P(x) \) pour l'intervalle \( [t_i, t_{i+1}] \) est construit de telle sorte qu'il coïncide avec la fonction \( f \) aux trois points d'interpolation :

1. \( P(t_i) = f(t_i) \)
2. \( P(m_i) = f(m_i) \)
3. \( P(t_{i+1}) = f(t_{i+1}) \)

où \( m_i = \frac{t_i + t_{i+1}}{2} \) est le point milieu de l'intervalle.

Le polynôme de Lagrange est donné par :

\[ P(x) = f(t_i) \frac{(x - m_i)(x - t_{i+1})}{(t_i - m_i)(t_i - t_{i+1})} + f(m_i) \frac{(x - t_i)(x - t_{i+1})}{(m_i - t_i)(m_i - t_{i+1})} + f(t_{i+1}) \frac

Le graphique montre la fonction \( f(t) = \sin(t) \) et son polynôme interpolateur \( P(x) \) sur l'intervalle \([0, \pi]\), utilisant la méthode de Simpson. Les trois points rouges indiquent les points d'interpolation où le polynôme coïncide exactement avec la fonction \( f \) :

1. À l'extrémité gauche de l'intervalle \( t_i \).
2. Au milieu de l'intervalle \( m_i \).
3. À l'extrémité droite de l'intervalle \( t_{i+1} \).

Le polynôme interpolateur de Lagrange \( P(x) \) est construit de manière à ce qu'il passe par ces trois points, capturant ainsi la tendance de la fonction \( f \) sur l'intervalle de manière plus précise qu'une approximation linéaire (comme dans la méthode des trapèzes) ou constante (comme dans la méthode des rectangles).

La méthode de Simpson est très efficace car elle prend en compte la courbure de la fonction et offre une approximation d'ordre supérieur par rapport aux méthodes des rectangles ou des trapèzes, ce qui se traduit généralement par une meilleure précision pour un nombre donné de sous-intervalles.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and the interval
def f(t):
    return np.sin(t)

a, b = 0, np.pi  # Define the interval
t_i = a
t_ip1 = b
m_i = (t_i + t_ip1) / 2  # Midpoint

# Define the Lagrange polynomials
def L0(x):
    return (x - m_i) * (x - t_ip1) / ((t_i - m_i) * (t_i - t_ip1))

def L1(x):
    return (x - t_i) * (x - t_ip1) / ((m_i - t_i) * (m_i - t_ip1))

def L2(x):
    return (x - t_i) * (x - m_i) / ((t_ip1 - t_i) * (t_ip1 - m_i))

# Define the interpolating polynomial
def P(x):
    return f(t_i) * L0(x) + f(m_i) * L1(x) + f(t_ip1) * L2(x)

# Create an array of points for plotting
x = np.linspace(a, b, 1000)
f_x = f(x)
P_x = P(x)

# Plot the function and the interpolating polynomial
plt.figure(figsize=(10, 6))
plt.plot(x, f_x, label='f(t) = sin(t)')
plt.plot(x, P_x, '--', label='Interpolating Polynomial P(x)')
plt.scatter([t_i, m_i, t_ip1], [f(t_i), f(m_i), f(t_ip1)], color='red', zorder=5)
plt.legend()
plt.xlabel('t')
plt.ylabel('f(t) and P(x)')
plt.title('Function and its Interpolating Polynomial using Simpson\'s Method')
plt.grid(True)
plt.show()
````

### Calcul primitive

````python
def integrate(f, a, b, N, method='midpoint'):
    """
    Calculate the numerical integral of the function f on the interval [a, b]
    using the specified method with N subintervals.

    Parameters:
    f (function): The function to integrate.
    a (float): The start of the integration interval.
    b (float): The end of the integration interval.
    N (int): The number of subintervals to use in the approximation.
    method (str): The method of numerical integration to use.
                  Options are 'left', 'right', 'midpoint', 'trapezoid', 'simpson'.

    Returns:
    float: The approximate integral of the function.
    """
    delta_x = (b - a) / N
    if method == 'left':
        return sum(f(a + i*delta_x) for i in range(N)) * delta_x
    elif method == 'right':
        return sum(f(a + (i+1)*delta_x) for i in range(N)) * delta_x
    elif method == 'midpoint':
        return sum(f(a + (i+0.5)*delta_x) for i in range(N)) * delta_x
    elif method == 'trapezoid':
        return (f(a) + 2 * sum(f(a + i*delta_x) for i in range(1, N)) + f(b)) * (delta_x / 2)
    elif method == 'simpson':
        x = np.linspace(a, b, N+1)
        y = f(x)
        return (delta_x/3) * (y[0] + 2 * sum(y[2:-1:2]) + 4 * sum(y[1::2]) + y[-1])
    else:
        raise ValueError("Unknown method: {}".format(method))

# Example usage:
# Define the function to integrate
def f_example(t):
    return t**3

# Calculate the integral using the midpoint method
integral_midpoint = integrate(f_example, 0, 2, 100, method='midpoint')
print("Midpoint Method: ", integral_midpoint)
````

En mathématiques, une fonction primitive \( F \) d'une fonction \( f \) est une fonction dont la dérivée est \( f \). C'est-à-dire que pour toute fonction \( f \) continue sur un intervalle, il existe une fonction \( F \) telle que \( F'(x) = f(x) \). Si vous avez une fonction primitive \( F \) qui s'annule en \( a \), cela signifie que \( F(a) = 0 \), et \( F \) est définie à une constante près. Pour n'importe quelle fonction continue \( f \), la fonction primitive qui s'annule en \( a \) peut être exprimée en utilisant l'intégrale définie comme suit :

\[ F(x) = \int_a^x f(t) \, dt \]

Cette définition garantit que \( F(a) = 0 \) car l'intégrale d'une fonction sur un intervalle de longueur nulle est nulle.

Pour calculer \( F \) pour différentes valeurs de \( x \) dans une boucle jusqu'à un certain \( t_{\text{max}} \), vous pouvez utiliser une somme cumulative des valeurs de \( f \) en utilisant une méthode d'intégration numérique. Chaque étape incrémentale dans la boucle ajoute l'aire sous la courbe \( f \) entre \( x \) et \( x + \Delta x \) à la somme cumulative précédente.

Voici

````python
def primitive(f, a, t_max, N):
    """
    Calculate the antiderivative F of the function f that vanishes at a.
    Loop over the interval [a, t_max] using N subintervals.

    Parameters:
    f (function): The function to integrate.
    a (float): The point where the antiderivative vanishes.
    t_max (float): The end of the interval for which to calculate the antiderivative.
    N (int): The number of subintervals to use in the approximation.

    Returns:
    list of tuples: Each tuple contains a value of t and the corresponding value of F(t).
    """
    delta_t = (t_max - a) / N
    t_values = np.linspace(a, t_max, N+1)
    F_values = [0]  # F(a) = 0 by definition

    # Calculate the antiderivative at each point
    for i in range(1, len(t_values)):
        # Approximate the integral from a to t_values[i] using the trapezoidal rule
        F_values.append(integrate(f, a, t_values[i], i, method='trapezoid'))
    
    return list(zip(t_values, F_values))

# Define the function to integrate
def f(t):
    return np.sin(t)

# Calculate the antiderivative of sin(t) that vanishes at 0 over [0, 2*pi
primitive(f,0,5,50)
````

Pour reconstruire le profil de vitesse et de position d'une boule de pétanque en chute libre depuis une hauteur de \( h = 100 \) mètres sans vitesse initiale et en négligeant les frottements, nous pouvons utiliser les lois de la cinématique. La boule de pétanque subit une accélération constante vers le bas due à la gravité, et nous pouvons utiliser les équations suivantes :

1. Équation de la position en fonction du temps :
\[ y(t) = h - \frac{1}{2} g t^2 \]

2. Équation de la vitesse en fonction du temps :
\[ v(t) = -gt \]

où :
- \( y(t) \) est la position verticale de la boule par rapport à sa position initiale à l'instant \( t \).
- \( h \) est la hauteur initiale depuis laquelle la boule tombe (\( h = 100 \) mètres dans ce cas).
- \( g \) est l'accélération due à la gravité (\( g \approx 9.81 \) m/s² sur Terre).
- \( v(t) \) est la vitesse de la boule à l'instant \( t \).

Nous pouvons utiliser ces équations pour calculer la position et la vitesse de la boule à différents moments dans sa chute en utilisant la méthode de notre choix (par exemple, la méthode du point milieu ou la méthode de Simpson).

Voyons comment cela peut être fait en utilisant la méthode du point milieu pour diviser le temps en intervalles et calculer la position et la vitesse à chaque étape. Nous prendrons un intervalle de temps \( \Delta t \) pour effectuer les calculs. La masse de la boule (\( m \)) n'affecte pas la cinématique de sa chute libre, elle n'entre donc pas en jeu dans ces calculs.

Voici le code Python pour calculer le profil de vitesse et de position de la boule de pétanque en chute libre :

Les graphiques ci-dessus représentent le profil de vitesse et de position d'une boule de pétanque en chute libre depuis une hauteur de 100 mètres sans vitesse initiale, en négligeant les frottements.

- Dans le graphique de la position par rapport au temps (à gauche), nous pouvons voir que la boule commence à une hauteur de 100 mètres (position initiale) et tombe progressivement vers le bas sous l'effet de la gravité. La position diminue paraboliquement avec le temps, comme prévu pour un objet en chute libre.

- Dans le graphique de la vitesse par rapport au temps (à droite), nous constatons que la vitesse de la boule augmente linéairement en magnitude (car la vitesse est négative, ce qui signifie qu'elle est dirigée vers le bas) en raison de l'accélération constante de la gravité. La vitesse atteint une valeur maximale (négative) lorsque la boule touche le sol.

Ces graphiques montrent comment la boule de pétanque se comporte au cours de sa chute libre, en suivant les lois de la cinématique. La méthode de calcul numérique (méthode du point milieu) nous permet de discrétiser le temps en intervalles et de calculer la position et la vitesse à chaque étape.

````python
# Define constants
h = 100  # Initial height in meters
g = 9.81  # Acceleration due to gravity in m/s^2
m = 0.73  # Mass of the boule in kg (730 g)

# Define time parameters
t_initial = 0  # Initial time
t_final = np.sqrt(2 * h / g)  # Time it takes to reach the ground (ignoring air resistance)
N = 1000  # Number of subintervals
delta_t = (t_final - t_initial) / N  # Time step

# Initialize arrays to store time, position, and velocity values
time_values = np.zeros(N+1)
position_values = np.zeros(N+1)
velocity_values = np.zeros(N+1)

# Perform numerical integration to calculate position and velocity at each time step
for i in range(N+1):
    time = t_initial + i * delta_t
    position = h - 0.5 * g * time**2
    velocity = -g * time
    
    # Store values in arrays
    time_values[i] = time
    position_values[i] = position
    velocity_values[i] = velocity

# Plot the profiles of position and velocity
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(time_values, position_values)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Position vs. Time for a Falling Boule")

plt.subplot(1, 2, 2)
plt.plot(time_values, velocity_values)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity vs. Time for a Falling Boule")

plt.tight_layout()
plt.show()
````
